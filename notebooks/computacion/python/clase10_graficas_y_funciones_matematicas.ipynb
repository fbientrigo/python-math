## Compresion de Lista
Es definir una Lista del tipo `x = [0,1,2,3]`, es escribir el siguiente codigo:

```python
x = []
for i in range(10): #i valdra 0, despues 1, despues 2,..., hasta 10-1 (osea 9)
    x.append(i) #agregamos el numero i a la lista

print(x) #veamos como quedo
```

Podemos hacer esto de manera compacta, asi:

```python
x = [i for i in range(10)] #creamos una lista de 0,1,2..,9
    #el primer i es lo que va dentor de la lista, luego la estructura del for
print(x)
```

Puede ser util si queremos aplicar una funcion rapida, como por ejemplo, multiplicar por dos antes de guardarlo en la lista

```python
x = [2*i for i in range(10)]
print(x)
```

```python
# Si solo quieres guardar lo que sale del range esto es mas rapido,
# convertirlo en una lista

x = list(range(0,10))
print(x)
```

# Map
Como podemos elevar cada elemento de `x` al cuadrado?

```python
y = x**2 #obtendremos un error
```

La forma clasica vendria a utilizar un loop for:

```python
y = [] #aqui va cada numero de x al cuadrado
for e in x: #e es un elemento de x, primero es 0, luego 1, .., asi hasta e = ultimo elemento de x
    y.append( e * e ) # agregamos el output a la lista `y`

print(y) #veamos como quedo
```

Pero esto es escribir mucho, resulta que hay una forma de aplicar una operacion a cada elemento de una lista

`map(operacion, lista)`

```python
def f(x):
    return x*x #esta es la funcion que queremos aplicar

y = map(f, x) #mira que solo escribimos el nombre de la funcion
y = list(y)  #al comienzo map nos da un objeto que no es una lista, es necesario convertirlo

print(y)
```

# Calculo Diferencial
Aqui comenzamos la parte de trabajar con Funciones y visualizar, ya que te ense√±e los trucos ahora podemos utilizarlos para hacer todo mas legible y rapido.

Comenzaremos aprendiendo a graficar una funcion basica

$$y = x^2 + 2x$$
con x de 0 a 10

```python
# Lo primero es crear la funcion que usaremos, le llamaremos f
def f(x):
    return x*x + 2*x

#Luego los x que veremos, 'la pre-imagen' o 'dominio'
x = [i for i in range(0,10+1)] #10+1 por que queremos que incluya el ultimo valor

#creamos el y, que seran la salida de la funcion f, o 'la imagen' o 'recorrido'
y = map(f,x) #aplicamos la funcion a cada x
y = list(y) #necesario al usar map, convertirlo en una lista

print('x:',x)
print('y:',y)
```

```python
#Ahora a graficar
plt.plot(x,y) #la funcion mas importante, es la que grafica
plt.show() #muestra el grafico creado
```

Es un buen grafico para partir, pero necesita que los ejes tengan nombres, estas funciones entre la creacion del grafico y mostrar el grafico seran perfectas

```python
plt.plot(x,y) 

plt.xlabel('x') #le da un nombre al eje x, u horizontal
plt.ylabel('y') #nombre al eje y, o vertical

plt.show()
```

```python
# Que tal escribir menos?
def graficar(x,y,nombrex,nombrey):
    plt.plot(x,y)
    plt.xlabel(nombrex)
    plt.ylabel(nombrey)
```

```python
#Repitamos el grafico aqui, aprovechare de agregarle un titulo
graficar(x,y,'x','y')
plt.title('Este es el titulo')
plt.show() #recuerda esta parte
```

Probemos ahora con una funcion matematica mas compleja, como un `Seno(x)`

```python
from math import sin #esto viene de las librerias que trae python

y2 = map(sin, x) #aplicamos la funcion a cada x
y2 = list(y2)

#Graficamos
graficar(x,y2,'x','seno(x)')
plt.show()
```
